# Top Player Examples

This notebook demonstrates simple SQLite queries against `data/latest.sqlite`.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../data/latest.sqlite')
```

```python
pd.read_sql_query('''
select player_name, team, top_speed_kmh
from player_physical_stats
where top_speed_kmh is not null
order by top_speed_kmh desc
limit 5
''', conn)
```

```python
pd.read_sql_query('''
select player_name, team, total_distance_m
from player_physical_stats
where total_distance_m is not null
order by total_distance_m desc
limit 5
''', conn)
```
