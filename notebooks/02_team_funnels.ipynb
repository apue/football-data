# Team Conversion Funnel

Compare possession, passing, progression, final-third receptions, shots, xG, and goals.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../data/latest.sqlite')
pd.read_sql_query('''
select m.match_no, t.team, t.possession_pct, t.passes_total,
       t.completed_line_breaks, t.receptions_final_third,
       t.attempts_total, t.xg, t.goals
from team_match_stats t
join matches m using(match_key)
order by m.match_no, t.team
''', conn)
```
